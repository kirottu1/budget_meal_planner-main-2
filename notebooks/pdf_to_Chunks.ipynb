```python
#Step One:
import pymupdf4llm
import pathlib
# Convert PDF to markdown
md_text = pymupdf4llm.to_markdown("../data/city_by_law/zoning_2021-08-03.pdf", page_chunks=False)
#Save markdown to file (optional) perhaps just save as a string
pathlib.Path("../data/city_by_law/zoning_2021-08-03.md").write_bytes(md_text.encode())
```

```python
#Step Two:
from langchain_text_splitters import MarkdownHeaderTextSplitter
# Define headers to split on
headers_to_split_on = [ ("#", "Header 1"), ("##", "Header 2"), ("###", "Header 3"), ]
# Initialize splitter
markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on)
# Split the markdown text
md_header_splits = markdown_splitter.split_text(md_text)
```

```python
import pymupdf # imports the pymupdf library
doc = pymupdf.open("../data/recipes/2022CookingAroundtheWorldCookbook.pdf") # open a document
extracted_text = []
for page in doc: # iterate the document pages
  extracted_text.append(page.get_text()) # get plain text encoded as UTF-8

print(len(extracted_text))
for text in extracted_text[13].split("\n"):
  print(text)
```
